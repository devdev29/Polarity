```python
import re 
import spacy
import contractions as cont
from bs4 import BeautifulSoup
import pandas as pd

def remove_repeat(text):
    newstr=''
    for word in str(text).split():
        if len(set(word))>1:
            newstr+=word+' '
    return newstr

def clean(text):
    lemmatizer=spacy.load('en_core_web_sm', disable=['parser', 'ner'])

    text=text.str.lower()
    text=text.transform(func=lambda x: re.sub('@\w*','',x))
    text=text.transform(func=lambda x:re.sub('http\S*','',x))
    text=text.transform(func=lambda x:re.sub('[0-9]','',x))
    text=text.transform(func=lambda x:re.sub(r"[',!%?/;:{}()\-\*\+\[\]\.\^]",'',x))
    text=text.transform(func=lambda x:BeautifulSoup(x,'lxml').get_text())
    text=text.transform(func=lambda x:cont.fix(x))
    text=text.transform(func=lambda x:''.join(word.lemma_+' ' for word in lemmatizer(x)))
    text=text.transform(func=lambda x:remove_repeat(x))
    return text

df= pd.read_csv('training.1600000.processed.noemoticon.csv')
cols = ['sentiment','id','date','query_string','user','text']
df=pd.read_csv('training.1600000.processed.noemoticon.csv', encoding='latin1',names=cols)
df=df[700000:900000]
df=df.drop(columns=['id','date','query_string','user'])
df['text']=clean(df['text'])
df.to_csv('s140-cleaned.csv',index=False)
df
```
